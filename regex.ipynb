```python
import re

s1 = 'Students IDs are 123 543 654 756'
s1_pattern = r"\d{3}"
reg = re.compile(s1_pattern)

res = reg.findall(s1)
print(res)

s2 = 'Sachin is cricket player and born in APR 24'
sac = re.match(r"\w{4}",s2)
print(sac)
print(sac.group())

pattern = r"\w{6}"
sac2 = re.search(pattern,s2)
print(sac2.group())

result = re.search(r"\d{2}$",s2)
print(result.group())

exact = re.findall(r"player",s2)
print(exact)

print(sac.start())
print(sac.end())

word = re.split(r"\s+",s2)
print(word)

num = "24-04-2023"
no = re.split(r"\D",num,maxsplit=3)
print(no)

deli = "23,54,75,32-54-65"
de = re.split(r"-|,",deli)
print(de)

new_s = "Python.com and it is; for, 5Python-developer"
ne = re.split(r"[-;,.\s]\s*",new_s)
print(ne)

ex = re.split(r"and[\s,-]+",new_s)
print(ex)

print(re.split('[A-Z]+',new_s))
print(re.split('[a-z]+',new_s, flags=re.IGNORECASE))

repl = re.sub(r"\s","_",new_s)
print(repl)

repl = re.sub(r"\s","",new_s)
print(repl)

repl = re.sub(r"\s","-",new_s, count=1)
print(repl)

subn = re.subn(r"[a-z]{2,}","FITA", new_s)
print(subn)
```
