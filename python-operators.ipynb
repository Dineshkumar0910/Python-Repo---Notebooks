```python
#Arithmetic operators (+,-,*,/,%)
n1 = 76
n2 = 64
print(n1+n2)
print(n1-n2)
print(n1*n2)
print(n1/n2)
print(n1%n2)
```

```python
print(n1**2)
```

```python
print((n1 + 5) * n2)
```

```python
#relational - (<,<=,>,>=,==,!=)
```

```python
print(n1<n2)
print(n1>n2)
print(n1<=n2)
print(n1>=n2)
print(n1==n2)
print(n1!=n2)
```

```python
#Assignment operators (=,+=,-=,*=,/=,%=)
```

```python
n1 += 5
print(n1)
```

```python
n1 -= 4
print(n1)
```

```python
n1 /= 2
print(n1)
```

```python
#membership operator - in
li = [4,7,1,5]
if 2 in li:
    print("yes")
else:
    print("no")
```

```python
#identity operator
x = 5
y = 5
print(x is not y)
```

```python
#logical operators - and,or,not
print(x and y)
print(x and 7)
```

```python
print( x or y)
```
