```python
class Cricket:
    def __init__(self,name,team,skill):
        self.name = name
        self.team = team
        self.skill = skill
    def IPL(self):
        print("Name",self.name,"Team",self.team,"Specialist",self.skill)

c = Cricket("Virat","RCB","Batting")
c.IPL()
```

```python
class Student:
    sch_name = "SBOA"
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
    def show(self):
        print("Student details:",self.name,self.age,Student.sch_name)
    
    @classmethod
    def school(self,sch_name):
        self.sch_name = sch_name
        
stu = Student("Rahul",12)
stu.show()

Student.school("New Public School")
stu.show()

del stu
print(stu)
```
