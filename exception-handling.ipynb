```python
def Calc(x,y):
    try:
        res = x // y
        print("Answer is:",res)
    except ZeroDivisionError:
        print("Not divided by Zero")

Calc(5,0)
```

```python
def fun(a,b):
    try:
        c = ((a+b)//(a-b))
    except ZeroDivisionError:
        print("Not divisible by 0")
    else:
        print(c)
    finally:
        print("This executes always")

fun(2,3)
fun(2,2)
```

```python
def fact(x):
    f = []
    for i in range(1, x+1):
        if x % i == 0:
            f.append(i)
    return f

n = 6008514475143
print(fact(n))
```
