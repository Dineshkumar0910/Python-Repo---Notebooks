```python
color = {'black','white','pink','beige','blue','gold','red'}
rainbow = {'violet','indigo','blue','green','yellow','red'}
```

```python
uni = color.union(rainbow)
print(uni)
```

```python
inter = color.intersection(rainbow)
print(inter)
```

```python
diff1 = color.difference(rainbow)
print(diff1)
diff2 = rainbow.difference(color)
print(diff2)
```
