```python
class Emp:
    def __init__(self,name,salary):
        self.name = name
        self.salary = salary
    def details(self):
        print("Name:",self.name,"Salary:",self.salary)

em = Emp('Sathish',120000)
print(em.name,em.salary)
em.details()
```

```python
class Company:
    def __init__(self):
        self._project = 'Python'

class Employee(Company):
    def __init__(self,name):
        self.name = name
        Company.__init__(self)
        
    def det(self):
        print("Emp name:",self.name)
        print("Group is",self._project)
        
emp = Employee('Jessie')
emp.det()
```

```python
class Stud:
    def __init__(self,name,age):
        self.name = name
        self.__age = age
        
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        self.__age = age
        
st = Stud('Karthik',21)
print(st.name,st.get_age())

st.set_age(23)
print(st.name,st.get_age())
```

```python
def gen():
    yield 1
    yield 2
    yield 3
    
for x in gen():
    print(x)
```

```python
def Num(n):
    for i in range(n):
        if(i%2!=0):
            yield i
res = Num(8)
print(next(res))
print(next(res))
print(next(res))
print(next(res))
print(next(res))
```

```python
def decr(fn):
    def wrap():
        print("before decrators")
        fn()
        print("after decrators")
    return wrap

def my_decr():
    print("this is decrators program")
    
my_decr = decr(my_decr)
my_decr()
```
