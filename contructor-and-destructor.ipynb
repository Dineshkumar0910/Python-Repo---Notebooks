```python
class Transport:
    def __init__(self,engine):
        self.engine = engine

class Car(Transport):
    def __init__(self,engine,speed):
        super().__init__(engine)
        self.speed = speed
        
T1 = Car("6 stroke", "120 kph")
print(f'Car speed is{T1.speed},Engine is{T1.engine}')
```

```python
class Car(Transport):
    def __init__(self,engine,speed):
        super().__init__(engine)
        self.speed = speed
    def details(self):
        print(f'Car speed is{T1.speed},Engine is{T1.engine}')
    def __del__(self):
        print("Object deleted")
        
T1 = Car("6 stroke", "120 kph")
T1.details()

del T1
```
